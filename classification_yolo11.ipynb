```python
from ultralytics import YOLO

# Load a COCO-pretrained YOLO11s-cls model
model = YOLO("YOLO11s-cls.pt")

# Train the model
results = model.train(data="custom_dataset", epochs=10, imgsz=640)

# FileNotFoundError: [Errno 2] No such file or directory: 'YOLO11s-cls.pt' 
#Download the pretrained weights from https://github.com/ultralytics/assets/releases/download/v8.3.0/yolo11s-cls.pt and paste it in your working directory
```
